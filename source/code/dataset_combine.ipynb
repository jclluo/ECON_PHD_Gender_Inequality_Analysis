```python
full_df = pd.concat([combined_data, missing_data])
full_df.head()
```

```python
# Convert all gender values to lowercase
full_df['gender'] = full_df['gender'].str.lower()

# Replace mostly_male with male and mostly_female with female
full_df['gender'] = full_df['gender'].replace('mostly_male', 'male')
full_df['gender'] = full_df['gender'].replace('mostly_female', 'female')

# Split 'andy' randomly between male and female
andy_mask = full_df['gender'] == 'andy'
andy_count = andy_mask.sum()
male_count = andy_count // 2
female_count = andy_count - male_count

full_df.loc[andy_mask, 'gender'] = ['male'] * male_count + ['female'] * female_count

# Show unique values after cleaning
full_df['gender'].unique()
```

```python
# Remove rows where gender is unknown
full_df = full_df[full_df['gender'] != 'unknown']
full_df[full_df['gender'] == 'unknown'] 
```

```python
# Check for missing values in placement column
# Remove rows with missing placement values
full_df = full_df.dropna(subset=['placement'])

# Remove duplicate rows
full_df = full_df.drop_duplicates()

print("After removing missing placements and duplicates:")
print(f"Number of rows: {len(full_df)}")
```
